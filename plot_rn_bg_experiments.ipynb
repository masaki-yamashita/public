```python
#XENON10   59+-2 uBq/kg Astroparticle Physics 34 (2011) 679–698 
#XENON100  33.4+-1.3 uBq/kg Eur. Phys. J. C (2017) 77:358
#PandaX-II 8.6+-4.6 uBq/kg   PRL 117, 121303 (2016)
#LUX   214Pb 3.5-3Hz/270 kg -> 12.9+-1  uB/kg Physics Procedia 61 (2015) 658 – 665
# XMASS-I 8.2+-0.5 mBq/835 kg  Nuclear Instruments and Methods in Physics Research A 716 (2013) 78–85
# XENON1T arXiv:2009.13981v3


# darwin arXiv:1309.7024v2
exp_name = ['XENON10 ','XENON100','LUX        ', 'PandaX-II ', 'XMASS-I ', 'XENON1T ', '', 'XENONnT Goal', 'DARWIN?']
mass = np.array   ([14, 62,   270 , 500, 832, 2000, 2000, 5900, 40000])
rate_rn = np.array([59, 33.4,  12.9, 8.6, 9.8, 12.7,  4.5,    1, 1])
rate_e  = np.array([2,  1.3  , 1.0 , 4.6, 0.6,  0.5,  0.1,    0, 0])
```

```python
plt.figure(figsize=(10, 6))
#plt.subplot(2,1,1)


plt.errorbar(mass, rate_rn,yerr=rate_e
             ,color='black',marker=".",ms=15, label="", elinewidth=2, capthick=2, capsize=2)
plt.xlim(10,100000)
plt.ylim(0.1,1000)
plt.xlabel("LXe Mass [kg]")
plt.ylabel("Rn [$\mu$ Bq/kg]")
plt.semilogy()
plt.semilogx()
plt.grid()

x_mass = np.linspace(10,100000, 1000)
y_rn = (x_mass)**(2/3.)*130/x_mass
plt.plot(x_mass, y_rn,"r--",lw=2, label="surf/volume: $x^{(2/3)}/x$")
plt.legend(loc='lower left')

for i in range(len(exp_name)):
    if i == len(exp_name)-2:        
        offset = 200
    else:
        offset = 20
    plt.text(mass[i], rate_rn[i]*offset, exp_name[i], size=18, rotation=90.,
             ha="center", va="top", color='darkgreen'
             )

```
