```python
exp_name = ['XENON10 ','XENON100','LUX', 'PandaX-II ', 'XMASS-I ', 'XENON1T ', '', 'XENONnT-SR0-2023', 'LZ-2023', 'DARWIN Goal']
mass = np.array   ([14, 62,   270 , 500, 832, 2000, 2000, 5900,7000, 40000])
rate_rn = np.array([59, 33.4,  12.9, 8.6, 9.8, 12.7,  4.5,    1.31, 3.26, 0.1])
rate_e  = np.array([2,  1.3  , 1.0 , 4.6, 0.6,  0.5,  0.1,    0.17, 0, 0])
```

```python
plt.figure(figsize=(12, 8))
#plt.subplot(2,1,1)


plt.errorbar(mass, rate_rn,yerr=rate_e
             ,color='black',marker=".",ms=12, label="", elinewidth=2, capthick=2, capsize=2)
plt.xlim(10,100000)
plt.ylim(0.01,1000)
plt.xlabel("LXe Mass [kg]")
plt.ylabel("Rn [$\mu$ Bq/kg]")
plt.semilogy()
plt.semilogx()
plt.grid()

x_mass = np.linspace(10,100000, 1000)
y_rn = (x_mass)**(2/3.)*130/x_mass
plt.plot(x_mass, y_rn,"r--",lw=2, label="surf/volume: $x^{(2/3)}/x$")
plt.legend(loc='lower left')

for i in range(len(exp_name)):
    if i == len(exp_name)-3:        
        offset = 250
    else:
        offset = 19
    plt.text(mass[i], rate_rn[i]*offset, exp_name[i], size=16, rotation=90.,
             ha="center", va="top", color='blue')

```
